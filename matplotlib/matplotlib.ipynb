```python
months = ["jan" , "feb" , "march" , "may" ]
sales = [30,40,20,60]
plt.plot(months,sales, color = "black" , linestyle="--" , linewidth=2 , marker = "<" , label = "monthly sales") 

# color = "black" / color = "k"   # 'k' stands for black in RGB (like 'r' = red, 'g' = green) / color = "#000000"   # Hexadecimal for black / 
# color = (0, 0, 0) / color = "0.0"   # String representing black (0 = black, 1 = white)

# marker="."   # Point
# marker="o"   # Circle
# marker="s"   # Square
# marker="^"   # Triangle up
# marker="<"   # Triangle left
# marker=">"   # Triangle right
# marker="v"   # Triangle down
# marker="D"   # Diamond
# marker="*"   # Star
# marker="x"   # X shape
# marker="+"   # Plus sign

#linestyle = "-." / "--" / "-"

plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend(loc="upper center" , fontsize=8)  #loc ="upper left" / "best" / "lower left" /"upper right" / "center left" / "center right" / "center"
                                            #/ "lower right" / "upper center" / "lower center" / "right"
plt.grid( color="gray" ,linestyle="--" , linewidth=1)
plt.xlim(0,2)
plt.ylim(0,70)
plt.xticks(["jan" , "feb" , "march" , "may" ],["m-1","m-2","m-3","m-4"])
plt.show()
```

```python
# plot barchart  ( X-axis (categories) , Y-axis (values) ) basic
# barchart with parameter plt.bar(X,height,width,bottom,align,color)
# x categories , height corresponding values , width width of the bar (default 0.8) , bottom baseline (default 0) , align : how to align bar (center or edge)
lables = ["mon","thue","web","tue","fri","stu","sun"]
sales = [20,40,10,30,50,90,60]
plt.bar(lables, sales)  #basic whith some default parameters
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales in Bar chart")
plt.show()
```

```python
plt.bar(lables, sales , width=0.3, bottom = 1, color = 'orange' ,align = 'edge', edgecolor='red' ) 
plt.show()
```

```python
#sub plot : multipal charts in one canvas
#usefull for comparison 
#subplot ( no rows , no column , index )
#Use plt.tight_layout() to avoid overlapping of subplots.
#Use plt.figure() to define the overall figure size/layout.


import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
data = [30, 44, 56, 76]

plt.figure(figsize=(10, 5))

# First subplot (Line plot)
plt.subplot(1, 2, 1)
plt.plot(x, data, color="black", linestyle="--", linewidth=2, marker="<", label="monthly sales")
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Sales")
plt.legend()

# Second subplot (Bar plot)
plt.subplot(1, 2, 2)
plt.bar(x, data, width=0.3, bottom=1, color='orange', align='edge', edgecolor='red')
plt.title("Bar Plot")
plt.xlabel("X-axis")
plt.ylabel("Sales")

plt.suptitle("Sub plot")
plt.tight_layout()
plt.show()
```

```python
# save fig ( file name with extention , dpi dots per inch (image resolution) , dbox inchs (fited plot in file) )


x = [1, 2, 3, 4]
data = [30, 44, 56, 76]

plt.plot(x, data, color="black", linestyle="--", linewidth=2, marker="<", label="monthly sales")
plt.title("Line Plot")
plt.legend()

# Save the figure
plt.savefig("sales_figure.png", dpi=300, bbox_inches='tight')
plt.show()
```
